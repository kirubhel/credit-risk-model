```python
## 1. Import Packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

%matplotlib inline
sns.set(style="whitegrid")
```

```python
## 2. Load Data
# Note: Place your dataset file in ../data/raw/data.csv
# For now, using a placeholder - replace with your actual data file
raw_path = "../data/raw/data.csv"

try:
    df = pd.read_csv(raw_path)
    print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
    display(df.head())
except FileNotFoundError:
    print(f"Data file not found at: {raw_path}")
    print("Please place your dataset in the data/raw/ directory")
    # Creating a sample dataframe for demonstration
    df = pd.DataFrame({
        'TransactionId': ['T001', 'T002', 'T003'],
        'Amount': [100.0, 250.5, 75.0],
        'FraudResult': [0, 1, 0]
    })
    print("Using sample data for demonstration")
    display(df.head())
```

```python
## 3. Data Types and Null Values
df.info()
df.isnull().sum().sort_values(ascending=False)
```

```python
## 4. Summary Statistics
df.describe()
```

```python
## 5. Distribution of Numerical Features
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols].hist(figsize=(15, 12), bins=30)
plt.suptitle("Numerical Feature Distributions")
plt.tight_layout()
plt.show()
```

```python
## 6. Distribution of Categorical Features
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    plt.figure(figsize=(10, 4))
    sns.countplot(y=col, data=df, order=df[col].value_counts().index)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
## 7. Correlation Matrix
plt.figure(figsize=(12, 10))
sns.heatmap(df[num_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
## 8. Boxplots for Outlier Detection
for col in num_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Outliers in {col}")
    plt.show()
```

```python
## 9. Missing Values
missing = df.isnull().mean().sort_values(ascending=False)
if missing[missing > 0].empty:
    print("No missing values found in the dataset!")
else:
    missing[missing > 0].plot(kind="bar", figsize=(10, 4), title="Missing Value Ratio")
    plt.show()
```
